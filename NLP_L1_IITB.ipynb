### $\textbf{Why is NLP So Heavily Dependent on Machine Learning?}$

1. **Ambiguity in Language**
   - **Phonetic Ambiguity**: Words sound the same but have different meanings (e.g., "write" vs. "right").
   - **Lexical Ambiguity**: Words have multiple meanings (e.g., "bank").
   - **Syntactic Ambiguity**: Sentence structure leads to multiple interpretations (e.g., "I saw the man with the telescope").
   - **Semantic Ambiguity**: Meaning changes with context (e.g., "I saw her duck").
   - **Pragmatic Ambiguity**: Social context alters intended meaning (e.g., "Can you pass the salt?").

2. **Complexity at Every Stage of Processing**
   - **Speech Recognition**: Converting spoken words to text.
   - **Tokenization**: Breaking text into individual words or phrases.
   - **Parsing**: Understanding grammatical structure.
   - **Semantic Analysis**: Interpreting meaning in context.
   - **Discourse Analysis**: Understanding sentence connections and narrative coherence.

3. **Machine Learning as a Solution**
   - **Learning from Data**: Understanding patterns from examples.
   - **Making Probabilistic Choices**: Calculating probabilities to determine the most likely interpretation.
   - **Feature Extraction**: Automatically identifying important features in data.
   - **Adaptability**: Training models for specific tasks, languages, or dialects.

4. **Examples of ML in NLP**
   - **Speech Recognition**: Handling accents and noise.
   - **Part-of-Speech Tagging**: Identifying grammatical parts in sentences.
   - **Named Entity Recognition**: Classifying names of people, places, or organizations.
   - **Machine Translation**: Translating text between languages using large datasets.

**Conclusion**: ML helps navigate the complexity of human language by learning from examples, making probabilistic choices, and adapting to contexts, enabling computers to understand and generate language.

### $\text{Deep Questions in Language}$

1. **Child Language Acquisition**
   - **Pattern Recognition**: Recognizing how sounds form words and sentences.
   - **Context Understanding**: Learning words and sentences within different contexts.
   - **Rule Formation**: Forming rules about grammar and structure autonomously.
   - **Challenges for ML**: Handling vast data and variation, learning from few examples, and understanding context deeply.

2. **Aphasia and Language Impairments**
   - **Understanding the Impact**: Studying how brain damage affects language.
   - **Developing Therapies**: Creating tools for diagnosis and therapy.
   - **Challenges for ML**: Adapting to individual differences, combining interdisciplinary knowledge, and developing practical, user-friendly applications.

**Conclusion**: Addressing these deep questions drives ML innovation, helping models learn better, understand context, and develop adaptive tools for language disorders, making NLP a crucial test bed for ML advancement.

### $\textbf{A Perspective on Machine Learning}$

1. **Table Lookup**:
   - **Description**: Uses a pre-stored table of questions and answers.
   - **Example**: Quickly answers "What is 5 + 3?" by looking up "5 + 3" and retrieving "8" from memory.

2. **Rules-Based (Statistical)**:
   - **Description**: Applies predefined rules to make decisions based on observed conditions.
   - **Example**: Uses rules like "If it's sunny, take sunglasses" or "If it's rainy, take an umbrella" to decide actions.

3. **Machine Learning: SVM, LR, etc.**:
   - **Description**: Learns patterns from data to make predictions or classifications.
   - **Example**: Trains on examples of spam and non-spam emails to classify new emails automatically.

4. **Deep Neural Networks (DNNs)**:
   - **Description**: Uses interconnected layers of neurons to understand complex patterns in data.
   - **Example**: Translates languages or recognizes faces by learning from large datasets, handling nuanced and intricate tasks effectively.

### $\textbf{Learning from Data with Probability-Based Scoring}$

In the discussion on leveraging data with probability-based scoring and reducing human dependency, deep learning emerges as a pivotal tool

- **Data + Classifier > Human Decision Maker!**
  - Integrating data with classifiers typically leads to more accurate decisions than relying solely on human judgment.

- **Advantages with Abundant Data:**
  - **High Precision:** Minimizes errors of commission—incorrect decisions.
  - **High Recall:** Minimizes errors of omission—missing crucial decisions.

- **Role of Human-Engineered Features:**
  - Effective learning heavily relies on well-defined features that capture essential characteristics of the data.

- **Deep Learning for Dependency Reduction:**
  - Deep learning facilitates end-to-end systems capable of autonomously learning essential properties at intermediate layers.
